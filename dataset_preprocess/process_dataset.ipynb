```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
# Check for duplicates in the 'english' column
english_duplicates = df[df['english'].duplicated(keep=False)]
print("Duplicates in 'english' column:")
print(english_duplicates)

# # Check for duplicates in the 'french' column
french_duplicates = df[df['french'].duplicated(keep=False)]
print("Duplicates in 'french' column:")
print(french_duplicates)
```

```python
import re

# Define a regex pattern for special characters (excluding basic punctuation and space)
pattern = r'[^\w .,!?\'"’–-“$€£¥₹%％”‘\t&*()\-+=@#<>:;\/\\\[\]{}|^~`…]'  # Added more characters


# Find rows to be removed
removed = df[df['english'].str.contains(pattern, regex=True) | df['french'].str.contains(pattern, regex=True)]

# Print the removed rows
print("Rows removed due to special characters:")
print(removed)
print(len(removed))

```

```python
# Apply the filter to keep only valid rows
mask = ~df['english'].str.contains(pattern, regex=True) & ~df['french'].str.contains(pattern, regex=True)
df = df[mask].reset_index(drop=True)
df
```

```python
# Regex pattern: any word with two consecutive capital letters
pattern = r'\b\w*[A-Z]{2,}\w*\b'

# Find rows to be removed
removed = df[df['english'].str.contains(pattern, regex=True) | df['french'].str.contains(pattern, regex=True)]

print("Rows removed due to consecutive capital letters:")
print(removed)
print(len(removed))
```

```python
# Remove those rows from the DataFrame
mask = ~df['english'].str.contains(pattern, regex=True) & ~df['french'].str.contains(pattern, regex=True)
df = df[mask].reset_index(drop=True)
df
```

```python
# Create mask for valid length ratio
mask = (
    (df['english'].str.len() < 2 * df['french'].str.len()) &
    (df['french'].str.len() < 2 * df['english'].str.len())
)

# Find and print rows to be removed
removed = df[~mask]
print("Rows removed due to sentence length ratio > 2:1:")
print(removed)
print(len(removed))
```

```python
# Apply the filter
df = df[mask].reset_index(drop=True)
df
```

```python
# Create mask for valid sentence length
mask = (
    (df['english'].str.split().str.len() > 5) & (df['english'].str.split().str.len() <= 100) &
    (df['french'].str.split().str.len() > 5) & (df['french'].str.split().str.len() <= 100)
)

# Print rows that will be removed
removed = df[~mask]
print("Rows removed due to sentence length (<5 or >100 words):")
print(removed)
print(len(removed))
```

```python
df = df[mask].reset_index(drop=True)
df
```

```python
def shift_punctuation(text):
    # Fix "word .word" -> "word. word" and "word ,word" -> "word, word"
    text = re.sub(r'\s+([.,])', r'\1', text)      # Remove space before punctuation
    text = re.sub(r'([.,])(\w)', r'\1 \2', text)  # Ensure space after punctuation if followed by a word
    return text
df['english'] = df['english'].apply(shift_punctuation)
df['french'] = df['french'].apply(shift_punctuation)
df
```

```python
def capitalize_standalone_i(text):
    # Replace ' i ' with ' I '
    # Also handle ' i.' ' i,' ' i!' etc. at the end of a sentence
    return re.sub(r'(?<=\s)i(?=\s|[.,!?;:])', 'I', text)
df['english'] = df['english'].apply(capitalize_standalone_i)
df['french'] = df['french'].apply(capitalize_standalone_i)
df
```
