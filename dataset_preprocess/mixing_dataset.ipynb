```python
import pandas as pd

MAX_LINES = 1000000

# Read up to 1 million lines from the French file
with open('OpenSubtitles.en-fr.fr', 'r', encoding='utf-8') as f_fr:
    french_sentences = [line.strip() for _, line in zip(range(MAX_LINES), f_fr)]

# Read up to 1 million lines from the English file
with open('filtered_data that can be used\sorted dataset by sources\europarl-v7.fr-en.en', 'r', encoding='utf-8') as f_en:
    english_sentences = [line.strip() for _, line in zip(range(MAX_LINES), f_en)]

# Make sure both have the same number of lines
assert len(french_sentences) == len(english_sentences), "Files are not aligned!"

# Create a DataFrame
df = pd.DataFrame({
    'french': french_sentences,
    'english': english_sentences
})

# Preview
df
```

```python
df3 = pd.read_csv(r"filtered_data that can be used\425k samples tatobae.tsv", sep='\t', names=["id", "french","id2","english"], encoding='utf-8')
df3 = df3[["english", "french"]]
import unicodedata

def normalize_text(text):
    if pd.isnull(text):
        return text
    # Lowercase, strip, and normalize unicode
    text = text.lower().strip()
    text = unicodedata.normalize('NFKC', text)
    return text

df3['english'] = df3['english'].apply(normalize_text)
df3['french'] = df3['french'].apply(normalize_text)
df3
```

```python
combined_df = pd.concat([df, df1, df2, df3], ignore_index=True)
combined_df
```
